```
from Bio import SeqIO

for record in SeqIO.parse('sequence.fasta', 'fasta'):
    print(record)
```

```
ncov_record = SeqIO.read('sequence.fasta', 'fasta')
ncov_record
```

```
ncov_dna = ncov_record.seq
ncov_dna
```

```
len(ncov_dna)
```

```
ncov_rna = ncov_dna.transcribe()
ncov_rna
```

```
ncov_prot = ncov_rna.translate()
ncov_prot
```

```
ncov_pep = ncov_prot.split('*')
ncov_pep
```

```
ncov_clean = [str(i) for i in ncov_pep]
ncov_clean
```

```
import pandas as pd
```

```
df = pd.DataFrame({'amino_acids':ncov_clean})
df
```

```
df['count'] = df['amino_acids'].str.len()
```

```
df.head()
```

```
df.nlargest(10, 'count')
```
